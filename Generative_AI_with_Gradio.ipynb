```python
from diffusers import StableDiffusionPipeline
import torch
#from accelerate import Accelerator

model = "runwayml/stable-diffusion-v1-5"
sd_pipeline = StableDiffusionPipeline.from_pretrained(model, torch_dtype=torch.float16)
sd_pipeline = sd_pipeline.to("cuda")
```

```python
# function to generate an image with the stable diffusion model
def image_generation(prompt):
    neg_prompt = """
    low quality, bad anatomy, bad hands, text, error, missing fingers, extra digit, fewer digits, cropped, worst quality, low quality
    """
    image = sd_pipeline(prompt, negative_prompt=neg_prompt).images[0]
    return image
```

```python
# the prompt
prompt = "an astronaut fishing in a river, standing by the river bank"

# generate the image
print(prompt)
generated_img = image_generation(prompt)
generated_img.save("./astronaut.png")
```

```python
import gradio as gr

# function to generate an image with the stable diffusion model
def image_generation(prompt):
    neg_prompt = """
    low quality, bad anatomy, bad hands, text, error, missing fingers, exaggerated features, incomplete image, extra digit, fewer digits, cropped, worst quality, bad quality
    """
    image = sd_pipeline(prompt, negative_prompt=neg_prompt).images[0]
    return image

generate_image = gr.Interface(fn=image_generation,
                              inputs=[gr.Textbox(label="Image Prompt")],
                              outputs=[gr.Image(label="Generated Image")],
                              title="Stable Diffusion Image Generator with Gradio. Created by Barine Sambaris",
                              description="Enter a prompt to generate an image with Stable Diffusion, and Gradio",
                              allow_flagging="never",
                              examples=["a cat running in the snow, twilight",
                                       "beautiful girl, walking on the runway, beauty pegeant"])

generate_image.launch()
```
